```python
import requests
from datetime import datetime

API_KEY = "YOUR_API_KEY_HERE"
BASE_URL = "https://api.openweathermap.org/data/2.5"

# --------------------------
# Current Weather
# --------------------------
def get_current_weather(city):
    url = f"{BASE_URL}/weather"

    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"
    }

    try:
        response = requests.get(url, params=params)

        if response.status_code == 200:
            return response.json()

        elif response.status_code == 404:
            print("City not found!")

        elif response.status_code == 401:
            print("Invalid API Key!")

        else:
            print("API Error:", response.status_code)

    except Exception as e:
        print("Error:", e)

    return None


# --------------------------
# Forecast
# --------------------------
def get_forecast(city):
    url = f"{BASE_URL}/forecast"

    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"
    }

    try:
        response = requests.get(url, params=params)

        if response.status_code == 200:
            return response.json()

    except Exception as e:
        print("Error:", e)

    return None


# --------------------------
# Display Current Weather
# --------------------------
def display_weather(data):

    print("\n🌤 WEATHER DASHBOARD")
    print("=" * 40)

    city = data["name"]
    country = data["sys"]["country"]

    temp = data["main"]["temp"]
    feels = data["main"]["feels_like"]

    humidity = data["main"]["humidity"]

    wind = data["wind"]["speed"]

    condition = data["weather"][0]["description"]

    sunrise = datetime.fromtimestamp(
        data["sys"]["sunrise"]
    ).strftime("%H:%M")

    sunset = datetime.fromtimestamp(
        data["sys"]["sunset"]
    ).strftime("%H:%M")

    print(f"📍 Location: {city}, {country}")
    print(f"🌡 Temperature: {temp}°C")
    print(f"🤗 Feels Like: {feels}°C")
    print(f"☁ Condition: {condition}")
    print(f"💧 Humidity: {humidity}%")
    print(f"💨 Wind Speed: {wind} m/s")
    print(f"🌅 Sunrise: {sunrise}")
    print(f"🌇 Sunset: {sunset}")


# --------------------------
# Display Forecast
# --------------------------
def display_forecast(data):

    print("\n📅 5-Day Forecast")
    print("=" * 40)

    forecasts = data["list"]

    shown_dates = set()

    for item in forecasts:

        date = item["dt_txt"].split()[0]

        if date not in shown_dates:

            shown_dates.add(date)

            temp = item["main"]["temp"]

            weather = item["weather"][0]["main"]

            print(
                f"{date} | {temp}°C | {weather}"
            )

            if len(shown_dates) == 5:
                break


# --------------------------
# Temperature Converter
# --------------------------
def celsius_to_fahrenheit(temp):
    return (temp * 9/5) + 32


# --------------------------
# Main Menu
# --------------------------
while True:

    print("\n")
    print("=" * 40)
    print(" WEATHER DASHBOARD ")
    print("=" * 40)

    print("1. Current Weather")
    print("2. 5-Day Forecast")
    print("3. Temperature Converter")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        city = input("Enter city name: ")

        weather_data = get_current_weather(city)

        if weather_data:
            display_weather(weather_data)

    elif choice == "2":

        city = input("Enter city name: ")

        forecast_data = get_forecast(city)

        if forecast_data:
            display_forecast(forecast_data)

    elif choice == "3":

        temp = float(
            input("Enter Celsius temperature: ")
        )

        print(
            f"{temp}°C = {celsius_to_fahrenheit(temp):.2f}°F"
        )

    elif choice == "4":

        print("Thank You!")
        break

    else:
        print("Invalid Choice!")
```
